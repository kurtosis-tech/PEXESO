```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
```

```python
# データの読み込み
equijoin_table = pd.read_csv("/nfs_p/datagusto/PEXESO/src/joinable_columns.csv")
pexeso_table = pd.read_csv("/nfs_p/datagusto/PEXESO/src/joinable_columns_pexeso.csv")

# Pexesoテーブルのパスをファイル名に変換し、.csvを取り除く
pexeso_table['Table1'] = pexeso_table['Table1'].apply(lambda x: os.path.splitext(os.path.basename(x))[0])
pexeso_table['Table2'] = pexeso_table['Table2'].apply(lambda x: os.path.splitext(os.path.basename(x))[0])

# ground truthと予測のセットを作成
equijoin_set = set((row['Table1'], row['Column1'], row['Table2'], row['Column2']) for _, row in equijoin_table.iterrows())
pexeso_set = set((row['Table1'], row['Column1'], row['Table2'], row['Column2']) for _, row in pexeso_table.iterrows())

# 真のポジティブ（Ground Truth）と予測結果のベクトルを作成
true_positive_vector = []
predicted_vector = []

# 全ての予測結果について検証
for prediction in pexeso_set:
    if prediction in equijoin_set:
        true_positive_vector.append(1)
        predicted_vector.append(1)
    else:
        true_positive_vector.append(0)
        predicted_vector.append(1)

# Ground Truthに存在し、予測されていない項目を追加
for ground_truth in equijoin_set:
    if ground_truth not in pexeso_set:
        true_positive_vector.append(1)
        predicted_vector.append(0)

# Precision, Recall, F1スコアを計算
precision = precision_score(true_positive_vector, predicted_vector)
recall = recall_score(true_positive_vector, predicted_vector)
f1 = f1_score(true_positive_vector, predicted_vector)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
import pandas as pd
import os
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

equijoin_threthold = 0.8 #この数値以上
pexeso_threthold = 0.01 #この数値以下

# データの読み込み
equijoin_table = pd.read_csv("/nfs_p/datagusto/PEXESO/src/joinable_columns.csv")
pexeso_table = pd.read_csv("/nfs_p/datagusto/PEXESO/src/joinable_columns_pexeso.csv")

# Pexesoテーブルのパスをファイル名に変換し、.csvを取り除く
pexeso_table['Table1'] = pexeso_table['Table1'].apply(lambda x: os.path.splitext(os.path.basename(x))[0])
pexeso_table['Table2'] = pexeso_table['Table2'].apply(lambda x: os.path.splitext(os.path.basename(x))[0])

equijoin_table['Intersection_Rate1'] = equijoin_table['Intersection_Rate1'].astype(float)
print(equijoin_table)

equijoin_table = equijoin_table[equijoin_table['Intersection_Rate1'] >= equijoin_threthold]

# スコアがn以上の予測結果をフィルタリング
filtered_pexeso_table = pexeso_table[pexeso_table['Score'] <= pexeso_threthold]

# ground truthと予測のセットを作成
equijoin_set = set((row['Table1'], row['Column1'], row['Table2'], row['Column2']) for _, row in equijoin_table.iterrows())
pexeso_set = set((row['Table1'], row['Column1'], row['Table2'], row['Column2']) for _, row in filtered_pexeso_table.iterrows())

# 真のポジティブ（Ground Truth）と予測結果のベクトルを作成
true_positive_vector = []
predicted_vector = []

false_positive = []

# 全ての予測結果について検証
for prediction in pexeso_set:
    if prediction in equijoin_set:
        true_positive_vector.append(1)
        predicted_vector.append(1)
    else:
        true_positive_vector.append(0)
        predicted_vector.append(1)
        false_positive.append(prediction)

# Ground Truthに存在し、予測されていない項目を追加
for ground_truth in equijoin_set:
    if ground_truth not in pexeso_set:
        true_positive_vector.append(1)
        predicted_vector.append(0)

# Precision, Recall, F1スコアを計算
precision = precision_score(true_positive_vector, predicted_vector)
recall = recall_score(true_positive_vector, predicted_vector)
f1 = f1_score(true_positive_vector, predicted_vector)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# コンヒュージョンマトリックスを計算
conf_matrix = confusion_matrix(true_positive_vector, predicted_vector)

# コンヒュージョンマトリックスをヒートマップとして表示
plt.figure(figsize=(4, 3))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Negative', 'Positive'], yticklabels=['False', 'True'])
plt.xlabel('Predicted')
plt.ylabel('Ground truth')
plt.title('')
plt.show()
```
